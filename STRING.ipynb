```python
prob = input("Enter: ")
dic = {}
quan = prob.replace(" ","")
for i in quan:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1
print(dic)
if dic[i] == 1:
  print("unique")
else:
  print("not unique")
max(dic,key=dic.get)
```

```python
give = input("enter the string: ")
take = int(input("enter the number: "))
start = give.split()
if take > len(start):
  print("Invalid INPUT")
else:
  start.pop(take-1)
  start.insert(take-1,"chumantar")
  end = " ".join(start)
  print(end)
```
