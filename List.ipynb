```python
#  Create a list of the first 10 Fibonacci numbers.
f = [0,1]
num = int(input("enter the number: "))
for i in range(1,num+1):
  a = f[i] + f[i-1]
  f.append(a)
print(f)
```

```python
# Rotate a list to the left by `n` positions.
# Using the existing 'all' list from the notebook state.
n = int(input("Enter the number of positions to rotate left: "))
original_list = f.copy() # Make a copy to preserve the original 'all' list
rotated_list = original_list[n:] + original_list[:n]
print(f"Original list: {original_list}")
print(f"Rotated list: {rotated_list}")
```

```python
# Create a list of prime numbers up to 50.
prime = []
# let's create a list first
for num in range(2,51):
  # we are creating two loops one for number and other for condition
   is_prime = True
  #  making the default condition or assumption as true
   for i in range(2,int(num**0.5)+1):
    if num%i == 0:
      is_prime = False
      break
    else:
      prime.append(num)
print(prime)
```

```python
# Split a list into chunks of size `n`
n = int(input("Enter the number : "))
chunk = [1,2,3,4,5,6,7,8,9,0,00,99,88,77,66,55,44,33,22,11]
chunks = [chunk[i:i+n] for i in range(0, len(chunk),n)]
print(chunks)
```

```python
# Find the second largest number in a list.
step =max(chunk)
chunk.remove(step)
ans = max(chunk)
print("Second larget in list of chunk is : ",ans)
chunk.append(step)
print("And the list is : \n",chunk)
```

```python
# Replace every element in a list with its square.
chunk = [i*i for i in chunk]
print(chunk)
```

```python
dic = {}
for k,v in enumerate(chunk):
  dic[v] = k
for r,a in dic.items():
  print(f"{r} : {a}")
```
