```python
lis = []
cll = int(input(print("Enter the amount you want to add : ")))
for i in range(cll):
  val = input(f"Enter the {i+1} value for list: ")
  lis.append(val)
print(lis)
n_lis = []
for terms in lis:
  if terms not in n_lis:
    n_lis.append(terms)
print("The new list without repetition : ",n_lis)
```
